### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px
import os
import warnings

warnings.filterwarnings("ignore")
```

```python
# helper function to calculate coupon acceptance rate
def get_coupon_usage_rate(data):
    return (data[data['Y'] == 1].shape[0] / data.shape[0]) * 100.0
```

```python
# For the column "toCoupon_GEQ5min" there is only  1 unique value for all the rows.
# Which says all the places in this are more than 5 mins away. 
# which is to say that all this column by itself is not a determining factor in accepting a coupon. 
# hence we can drop this column
df.drop(columns=['toCoupon_GEQ5min'], inplace=True)
df.isna().sum()
```

1. Create a new `DataFrame` that contains just the bar coupons.

2. What proportion of bar coupons were accepted?


```python
# calculating percent of total observations who accepted the bar coupon.
bar_coupon_used_percentage = get_coupon_usage_rate(bar_df)
print(f"\n The percentage of customers accepting Bar coupons is {bar_coupon_used_percentage}")

# plotting pie chart
plt.pie([bar_coupon_used_percentage, 100.0 - bar_coupon_used_percentage], 
        labels=['Used', 'Did Not Use'], autopct='%.2f%%') 
plt.title('Bar Coupon Usage')
plt.savefig('images/barQ2_coupon_usage.png')
plt.show() 
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
print(bar_df["Bar"].value_counts())

all_vals_3_or_less = ["never","less1","1~3"]
bar_df_all_vals_3_or_less = bar_df.query("Bar in @all_vals_3_or_less")
bar_df_rest_of_vals = bar_df.query("Bar not in @all_vals_3_or_less")

bar_coupon_lt_3_used_percent = get_coupon_usage_rate(bar_df_all_vals_3_or_less)
bar_coupon_gt_3_used_percent = get_coupon_usage_rate(bar_df_rest_of_vals)
print("Less than (or equal to) 3 visits: ", bar_coupon_lt_3_used_percent)
print("Greater than 3 visits: ", bar_coupon_gt_3_used_percent)
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# bar coupon and age categories to focus 
bar_morethan_once = ['1~3', '4~8', 'gt8']
age_morethan_25 = ['26', '31', '36', '41', '46', '50plus']

# getting group specific rows
bar_morethan_once_and_age_morethan_25 = bar_df.query("Bar in @bar_morethan_once and age in @age_morethan_25")
bar_otherthan_once_and_age_otherthan_25 = bar_df.query("Bar not in @bar_morethan_once and age not in @age_morethan_25")

# calculating coupon usage rate for the above groups
percent1 = get_coupon_usage_rate(bar_morethan_once_and_age_morethan_25)
percent2 = get_coupon_usage_rate(bar_otherthan_once_and_age_otherthan_25)

# saving percents for plotting
percents = [percent1, percent2]

# plotting percentages in bar plot with labels
plt.figure(figsize=(10, 6)) 
ax = sns.barplot(x=['Bar Coupon (vistis more than once and age GT 25', 'All Others'], y=percents)
plt.ylabel('Coupon Usage Rate')
plt.title('Coupon Usage Rate - Bar Goers More than once and Over the Age of 25 VS The Rest')
plt.savefig('images/BarQ4.png')
plt.show()
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
# bar coupon and age categories to focus 
bar_morethan_once = ['1~3', '4~8', 'gt8']

# getting group specific rows
bar_morethan_once_and_noFFF_and_nokids = bar_df.query("Bar in @bar_morethan_once and passanger != 'Kid(s)'and occupation != 'Farming Fishing & Forestry'")
bar_otherthan_once_and_FFF_and_kids = bar_df.query("Bar not in @bar_morethan_once and passanger == 'Kid(s)' & occupation == 'Farming Fishing & Forestry'")

# calculating coupon usage rate for the above groups
percent1 = get_coupon_usage_rate(bar_morethan_once_and_noFFF_and_nokids)
percent2 = get_coupon_usage_rate(bar_otherthan_once_and_FFF_and_kids)

# saving percents for plotting
percents = [percent1, percent2]

# plotting percentages in bar plot with labels
plt.figure(figsize=(10, 6)) 
ax = sns.barplot(x=['Bar More than once,no Kids,no Farming', 'Bar Less than once, Kids and Farming'], y=percents)
plt.ylabel('Coupon Usage Rate')
plt.title('Coupon Usage Rate - (Bar Goers More than once, no Kids, no Farming VS  Bar Goers Less than once, with Kids, and Farming')
plt.savefig('images/BarQ5.png')
plt.show()
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
# data categories
age_under_30 = ['21', '26', 'below21']
visits_4_or_more = ['4~8', 'gt8']
salary_less_than_50k = ['$37500 - $49999', '$12500 - $24999', '$25000 - $37499', 'Less than $12500']

# getting specific customer groups
bar_1m_nokid_widow = bar_df[(bar_df['Bar'].isin(bar_morethan_once)) & 
    (bar_df['passanger'] != 'Kid(s)') & 
    (bar_df['maritalStatus'] != 'Widowed')]
bar_1m_age_under_30 = bar_df[(bar_df['Bar'].isin(bar_morethan_once)) & 
    (bar_df['age'].isin(age_under_30))]
bar_cheap_salary_lessthan_50k = bar_df[(bar_df['RestaurantLessThan20'].isin(visits_4_or_more)) & 
    (bar_df['income'].isin(salary_less_than_50k))]

# calculating coupon usage rate for each customer group
percent1 = get_coupon_usage_rate(bar_1m_nokid_widow)
percent2 = get_coupon_usage_rate(bar_1m_age_under_30)
percent3 = get_coupon_usage_rate(bar_cheap_salary_lessthan_50k)

# saving percents for plotting
percents = [percent1, percent2, percent3]

plt.figure(figsize=(10,10))
# plotting multiple pie charts in one plot
ax1 = plt.subplot2grid((1, 3), (0, 0))
plt.suptitle('Coupon Usage Rates', y=0.8)
plt.pie([percent1, 100.0 - percent1], labels=['Yes', 'No'], autopct='%.2f%%')
plt.title('Bar Goers with Non-kid Passengers,Not Widowed', fontsize=8)

ax1 = plt.subplot2grid((1, 3), (0, 1))
plt.pie([percent2, 100.0 - percent2], labels=['Yes', 'No'], autopct='%.2f%%') 
plt.title('Bar Goers Aged 30 or younger', fontsize=8)

ax1 = plt.subplot2grid((1, 3), (0, 2))
plt.pie([percent3, 100.0 - percent3], labels=['Yes', 'No'], autopct='%.2f%%') 
plt.title('Regular Goers to Cheap Restaurants with Salary less than 50k', fontsize=8)
plt.savefig('images/BarQ6.png')
plt.show()
```

### Usage Rate of Coupon Types and Analysis of Coffee House Coupons 

```python
# Getting the data of different coupon types into independent dataframes.
# We already have the dataframe for bar coupons in bar_df. 
# Create the dataframe for coupon types: 'Coffee House', 'Carry Out & Take away', 'Restaurant(<20)','Restaurant(20-50)','Coffee House'
coffee_df = df[df['coupon'] == 'Coffee House']
takeout_df = df[df['coupon'] == 'Carry out & Take away']
rest20_df = df[df['coupon'] == 'Restaurant(<20)']
rest50_df = df[df['coupon'] == 'Restaurant(20-50)']

print (bar_df.shape)
print (coffee_df.shape)
print (takeout_df.shape)
print (rest20_df.shape)
print (rest50_df.shape)
```

```python
#calculate coupon usage percentage for different coupon types
bar_percent = get_coupon_usage_rate(bar_df)
coffee_percent = get_coupon_usage_rate(coffee_df)
takeout_percent = get_coupon_usage_rate(takeout_df)
rest20_percent = get_coupon_usage_rate(rest20_df)
rest50_percent = get_coupon_usage_rate(rest50_df)
```

```python
# saving percents for plotting
percents = [bar_percent, coffee_percent, takeout_percent, rest20_percent, rest50_percent]
labels = ['Bar', 'Coffee House','Take Out', 'Rest.(<20)', 'Rest.(20-50)']
# plotting percentages in bar plot with labels
ax = sns.barplot(x=labels, y=percents)
patches = ax.patches
for i in range(len(patches)):
   x = patches[i].get_x() + patches[i].get_width()/2
   y = patches[i].get_height()
   ax.annotate('{:.1f}%'.format(percents[i]), (x, y), ha='center')
plt.xlabel('Coupon Types')
plt.ylabel('Usage Rate')
plt.title('Usage Rate - Coupon Types')
plt.savefig('images/allcoupons.png')
plt.show()
```

```python
percent1_1d = get_coupon_usage_rate(bar_df[bar_df['expiration'] == '1d'])
percent2_1d = get_coupon_usage_rate(coffee_df[coffee_df['expiration'] == '1d'])
percent3_1d = get_coupon_usage_rate(takeout_df[takeout_df['expiration'] == '1d'])
percent4_1d = get_coupon_usage_rate(rest20_df[rest20_df['expiration'] == '1d'])
percent5_1d = get_coupon_usage_rate(rest50_df[rest50_df['expiration'] == '1d'])

percent1_2h = get_coupon_usage_rate(bar_df[bar_df['expiration'] == '2h'])
percent2_2h = get_coupon_usage_rate(coffee_df[coffee_df['expiration'] == '2h'])
percent3_2h = get_coupon_usage_rate(takeout_df[takeout_df['expiration'] == '2h'])
percent4_2h = get_coupon_usage_rate(rest20_df[rest20_df['expiration'] == '2h'])
percent5_2h = get_coupon_usage_rate(rest50_df[rest50_df['expiration'] == '2h'])

percents_1d = [percent1_1d, percent2_1d, percent3_1d, percent4_1d, percent5_1d]
percents_2h = [percent1_2h, percent2_2h, percent3_2h, percent4_2h, percent5_2h]
labels = ['Bar', 'Coffee House','Take Out', 'Rest.(<20)', 'Rest.(20-50)']
# plotting percentages in bar plot with labels
ax = sns.barplot(x=labels, y=percents_1d)
patches = ax.patches
for i in range(len(patches)):
   x = patches[i].get_x() + patches[i].get_width()/2
   y = patches[i].get_height()
   ax.annotate('{:.1f}%'.format(percents_1d[i]), (x, y), ha='center')
plt.xlabel('Coupon Types')
plt.ylabel('1-Day Coupon Usage Rate')
plt.title('Usage Rate - 1-Day Coupons')
plt.savefig('images/daycoupons.png')
plt.show()

# plotting percentages in bar plot with labels
ax = sns.barplot(x=labels, y=percents_2h)
patches = ax.patches
for i in range(len(patches)):
   x = patches[i].get_x() + patches[i].get_width()/2
   y = patches[i].get_height()
   ax.annotate('{:.1f}%'.format(percents_2h[i]), (x, y), ha='center')
plt.xlabel('Coupon Types')
plt.ylabel('2-hour Coupon Usage Rate')
plt.title('Usage Rate - 2-Hour Coupons')
plt.savefig('images/hourcoupons.png')
plt.show()
```
